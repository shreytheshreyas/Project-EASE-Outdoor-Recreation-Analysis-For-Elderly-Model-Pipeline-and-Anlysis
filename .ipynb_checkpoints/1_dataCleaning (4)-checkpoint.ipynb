```python
def loopFiles(folder,out, filetypes = ['.txt']):
    for file in listdir(folder):
        if file.startswith('._'):
            continue
            
        filePath = folder+"/"+file
        file_extension = os.path.splitext(filePath)[1]

        if(os.path.isdir(filePath)):
            loopFiles(filePath,out, filetypes)
        elif(file_extension in filetypes):
            out.append(filePath)
```

```python
def parseFilePath(path):

    [date,fileName] = (path.split("/"))
    [time,ext] = (fileName.split("."))

    
    row = {}
    
    row['date'] = date
    row['time'] = time
#     row['vidLength'] = vidLength
    return row
    
```

```python
def dotVis(df, bg_img, group_by = 'class', groups = None):

    if groups == None:
        groups = df[group_by].unique()
    
    plt.clf()
    cols = 4
    rows = math.ceil(len(groups)/4)
    n_axes = cols * rows
    (h,w,p) = bg_img.shape
    img_ratio = h/w

    

    s = 3
    fig, ax = plt.subplots(rows,cols, figsize=(cols*s, rows*s*img_ratio))
    colors = list(mcolors.TABLEAU_COLORS)

                
            
    for i,group in enumerate(groups):

        col = i % cols
        row = int((i - col)/cols)
        c = getColor(i)*255
        c = np.append(c, 255)
        
        if len(groups) < 4:
            sel_ax = ax[col]
        else:
            sel_ax = ax[row, col] 
            


        

        dots_object = df[(df['class'] == group)]
        
        dots_object = list(zip(dots_object.centroidX, dots_object.centroidY))
        im_dots = np.zeros((h,w,4), np.uint8)
        im_dots[:,:,-1] = (255)*0.5
        im_dots[:,:,0] = 255
        im_dots[:,:,1] = 255
        im_dots[:,:,2] = 255
#         print(im_dots)



        for x,y in dots_object:
            cv2.circle(im_dots, (x,y), radius=2, color=c,  thickness=-1)
            
        
            
        bg_img_= cv2.cvtColor(bg_img.copy(), cv2.COLOR_RGB2RGBA)


        sel_ax.imshow(bg_img_)
        sel_ax.imshow(im_dots)
        sel_ax.set_title(label=f'{objects_dict[group]}',loc='left',fontsize='small')

#     fig.tight_layout()


    
    # remove axis for all
    for row in range(rows):
        for col in range(cols):
            i = col + (row * cols)
            if len(groups) < 4:
                sel_ax = ax[col]
            else:
                sel_ax = ax[row, col] 
            sel_ax.axis('off')
            
    plt.show()
```

```python
def getColor(i):
    colors = list(mcolors.TABLEAU_COLORS)
    i = i%len(colors)
    return np.array(mcolors.to_rgb(list(mcolors.TABLEAU_COLORS)[i]))
```

```python
def barPlot(df, x, y, title,order = None,):
    plt.clf()
    
    if order is None: 
        order = df[x].unique()
        

    figsize = (3, len(order)*0.1+4)
    fig, ax = plt.subplots()
    
    bar = df.set_index(x).reindex(order)

    bar.plot.barh(
        figsize = figsize,
#             x=index, 
        y=y, 
        legend = False,
        ax = ax,
        color = [getColor(i) for i in range(20)],
        title = title,
        log = True
    )
    plt.show()
```

# Processing

## Extract video info and background images

```python

# get a list of all video files
video_paths = []
loopFiles(videos,video_paths,filetypes = ['.mp4'])
print('No. of video files:', len(video_paths))
video_paths.sort()
```

```python
# loop through videos list and extract information
videoFileCount = 0
videoData = []


extract_bg_images = True #flag to save bg_images


for file_path in tqdm(video_paths):
    video = os.path.splitext(file_path.replace(videos+"/",""))[0]
    images_path = f'{out_data}/bg/{video}'
    bg_img_path = f'{images_path}/bg_img.png'
    [date,time] = (video.split("/"))
    time = time.split("_")[0]
    
    
    if not (os.path.exists(images_path)):
        os.makedirs(images_path)

    try:
        vid_cap= cv2.VideoCapture(file_path)
        fps = vid_cap.get(cv2.CAP_PROP_FPS)
        frame_count = int(vid_cap.get(cv2.CAP_PROP_FRAME_COUNT))
        vid_width = int(vid_cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        vid_height = int(vid_cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        duration = int(frame_count/fps) if frame_count > 0 else 0

        
        if bg_path is None:
            success,image = vid_cap.read()
            if success:
                bg_img = image
                bg_path = f'{out_data}/bg_img.png'
                cv2.imwrite(bg_path, image,[int(cv2.IMWRITE_JPEG_QUALITY), 30])
            
        
        if extract_bg_images and not os.path.exists(bg_img_path):
            success,image = vid_cap.read()
        
            if success:
                cv2.imwrite(bg_img_path, image,[int(cv2.IMWRITE_JPEG_QUALITY), 30])

        vid_cap.release()
        
        videoData.append({
            'fps':fps, 
            'frame_count':frame_count,
            'vid_width': vid_width,
            'vid_height': vid_height,
            'duration': duration,
            'bg_img': bg_img_path,
            'video': video,
            'date':date,
            'time':time,
            'vid_id': videoFileCount,
            'file_path': file_path
            
            
        })
        
        videoFileCount +=1
    
    except Exception as e:
        print(e)
        continue
```

```python
# create dataframe
video_df = pd.DataFrame.from_dict(videoData)

video_df['timeStamp'] = pd.to_datetime(video_df['date']+video_df['time'], format="%y%m%d%H%M%S")
pd.to_datetime(video_df['timeStamp'])
video_df['timeStamp_end'] = video_df['timeStamp'] + pd.TimedeltaIndex(video_df['duration'], 's')
video_df
```

## Extract hourly thumbnails

```python
def sort_files_by_size(directory):
    # Get a list of all files in the directory
    files = os.listdir(directory)

    # Create a dictionary to store the file sizes
    file_sizes = {}

    # Loop through the files and get their sizes
    for file in files:
        path = os.path.join(directory, file)
        if os.path.isfile(path):
            size = os.path.getsize(path)
            file_sizes[file] = size

    # Sort the files by size in ascending order
    sorted_files = sorted(file_sizes.items(), key=lambda x: x[1], reverse = True)
    return sorted_files[0][0]
```

```python
video_df['timeStamp'].min()
video_df['timeStamp_end'].max()
```

```python
# extract_hourly_images = False
hourly_low_res = True
hourly_width = 600

t1 = None
for i,video in video_df.sort_values(by = "timeStamp").iterrows():
    hourly_height = int(hourly_width * (video['vid_height']/video['vid_width']))
    frames = []
    

    if t1 is None:
        frames.append({
            "frame": 0,
            "timeStamp": video.timeStamp,
            "time": 0
        })
    t1 = video.timeStamp.replace(minute=0, second=0)

    while t1< video.timeStamp_end:
        nextHr = t1 + timedelta(hours = 1)

        if nextHr < video.timeStamp_end:
            frames.append({
                "frame": (nextHr - video.timeStamp).total_seconds()*video.fps,
                "timeStamp": nextHr,
                "time": (nextHr - video.timeStamp).total_seconds()
            })

        t1 = nextHr
        
    if(len(frames) == 0):
        continue
        



    images_path = f'{out_data}/hourly'
    
    if hourly_low_res:
        images_path = f'{out_data}/hourly_small'
    if not (os.path.exists(images_path)):
        os.makedirs(images_path)



    if extract_hourly_images:

        for frame in frames:
            temp_folder = f'{images_path}/temp'


            if not (os.path.exists(temp_folder)):
                os.makedirs(temp_folder)


            cap = cv2.VideoCapture(video["file_path"])


            frameCount = 0

            nextFrame = frame["frame"]
            while(cap.isOpened()):
                cap.set(cv2.CAP_PROP_POS_FRAMES , nextFrame)
                ret, img = cap.read()
                if(ret):
                    img_path = f'{temp_folder}/{nextFrame+frameCount}.jpeg'
                    if hourly_low_res: 
                        img = cv2.resize(img,(hourly_width, hourly_height))
                    cv2.imwrite(img_path, img, [cv2.IMWRITE_JPEG_QUALITY, 50])
                frameCount+=1
                nextFrame += video.fps
                if((frameCount > 2) or (nextFrame > video["frame_count"])):
                    cap.release()
                    cv2.destroyAllWindows()
                    break


            sel_img = sort_files_by_size(temp_folder)
            try:
                shutil.move(f'{temp_folder}/{sel_img}', f'{images_path}/{frame["timeStamp"].strftime("%m-%d-%Y_%H%M%S")}.jpeg')
            except Exception as e:
                print(f"Error: {e}")


            try:
                shutil.rmtree(temp_folder)
            except Exception as e:
                print(f"Error: {e}")
    ;
```
