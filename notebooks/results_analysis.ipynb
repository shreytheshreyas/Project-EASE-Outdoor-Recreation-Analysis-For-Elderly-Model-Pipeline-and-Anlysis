## Normal BoxMot Tracking and Yolov8 Detection and their respective properties

The BoxMot Object tracking algorithm uses the regular YoloV8 model for the object detection. The model outputs a resuls object and each result is associated with a detection with the following properties: 
1. frameIdx.
2. bounding_box_id or object_id.
3. bounding_box_ltwh: xy coordinates of the top-left corner and bounding box widh and height.
4. classification: predicted classification by YOLOV8 algorithm.
5. confidence: Probabilty of correct classication. 

```python
sample_results_normal = pd.read_csv('./yolo_tracking/runs/track/exp59/footage/tampines_sample.mp4.csv')
column_headings_normal = ['frame_id', 'object_id', 'x-coordinate', 'y-coordindate','box_width','box_height', 'confidence', 'classification', 'Dont-Care', 'Dont-Care']
sample_results_normal.columns = column_headings_normal
sample_results_normal
```

## Pose Estmation Pipeline Integration along with bounding boxes and confidence Scores 

### Pose estimation details 
The YoloV8-pose estimation model outputs a results objects, similar to the other the other models with the exception that the results object has a KeyPoints Property. 
The Keypoints property is itself an object which has the following properties:
1. Confidence: Holds array of confidence scores for each detected keypoint.
2. data: Holds a multidimensional array of the spacial coordinates of the detected keypoints. Note that this is a three dimensional array and can hold spacial information for multiple object detections.
3. xy: Similar to data in a way that it holds a multi-dimensional array with exception to the coordinates on the z-axis.
4. xyn: Similar to xy property, but holds the normalized value of the coordinates.

The Keypoints property of the results object holds information for 17 respective points present in the body, which are in the order as follows: 
1. Nose
1. Left Eye
1. Right Eye
1. Left Ear
1. Right Ear
1. Left Shoulder
1. Right Shoulder
1. Left Elbow
1. Right Elbow
1. Left Wrist
1. Right Wrist
1. Left Hip
1. Right Hip
1. Left Knee
1. Right Knee
1. Left Ankle
1. Right Ankle
